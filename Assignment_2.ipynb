```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
housing = pd.read_csv('/content/Housing.csv')
housing
```

```python
housing.head()
```

```python
Housing1=housing.isnull().any()
Housing1
```

```python
Housing2=housing.isnull().sum()
print("no:of null values in each column:\n",Housing2)
```

```python
housing['price'].interpolate(method="linear",inplace=True) 
housing['area'].interpolate(method="linear", inplace=True)
housing['bedrooms'].interpolate(method="linear",inplace=True)
housing['bathrooms'].interpolate(method="linear",inplace=True)
housing ['stories'].interpolate (method="linear",inplace=True)
housing['parking'].interpolate (method="linear",inplace=True)
housing['mainroad'].fillna(method='backfill', axis = 0,inplace=True) 
housing ['guestroom'].fillna(method='backfill', axis = 0,inplace=True)
housing['basement'].fillna(method='backfill', axis = 0,inplace=True)
housing['hotwaterheating'].fillna(method='backfill', axis=0,inplace=True)
housing ['airconditioning'].fillna(method='backfill', axis=0,inplace=True) 
housing['prefarea'].fillna(method='backfill',axis=0,inplace=True)
housing['furnishingstatus'].fillna(method='backfill', axis = 0,inplace=True)
```

```python
housing3=housing.isnull().sum()
print("no:of null values:\n",housing3)
```
